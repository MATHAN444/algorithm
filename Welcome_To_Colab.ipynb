```
import random
import time
import matplotlib.pyplot as plt

# Insertion Sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# Heap Sort
def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[left] > arr[largest]:
        largest = left

    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    n = len(arr)

    # Build a max heap
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    # Extract elements from the heap
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)
    return arr

# Function to measure the time for sorting
def measure_sort_time(sort_function, n):
    arr = random.sample(range(1, 10 * n), n)  # Generate a list of n unique elements
    start_time = time.time()
    sort_function(arr)
    end_time = time.time()
    return end_time - start_time

# Experiment with different values of n
n_values = [10, 100, 500, 1000, 5000, 10000, 20000]
insertion_times = []
heap_times = []

for n in n_values:
    insertion_time = measure_sort_time(insertion_sort, n)
    heap_time = measure_sort_time(heap_sort, n)

    insertion_times.append(insertion_time)
    heap_times.append(heap_time)

# Plotting the time taken vs. number of elements for both algorithms
plt.plot(n_values, insertion_times, marker='o', label='Insertion Sort')
plt.plot(n_values, heap_times, marker='x', label='Heap Sort')
plt.xlabel('Number of elements (n)')
plt.ylabel('Time taken (seconds)')
plt.title('Sorting Time Comparison: Insertion Sort vs Heap Sort')
plt.legend()
plt.grid(True)
plt.show()
```
